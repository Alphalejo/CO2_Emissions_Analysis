```python
# Countries df = only countries info
# Region df = excludes the countries and only has the info from continents, subcontitnent and for income range
Countries_df = df[df["Code"].notnull()]
Region_df = df[df["Code"].isnull()].drop("Code", axis=1)
```

```python
Countries_df.head()
```

```python
# Some values werent measured so I replaced them for nan values to avoid it affecting the nmerical analysis 
Countries_df.replace([0.0, ""], np.nan, inplace=True)
df.replace([0.0, ""], np.nan, inplace=True)
```

```python
Countries_df.to_csv("../Data/Processed/Countries_df.csv", index=False)
Region_df.to_csv("../Data/Processed/Region_df.csv", index=False)
df.to_csv("../Data/Processed/worldwide_df.csv", index=False)
```
